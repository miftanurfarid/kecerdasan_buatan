```python
import numpy as np
import pygad
```

```python
def pers1(x):
    return x**2 + 2*x * 3
```

```python
x = np.arange(-5,5,.1)
```

```python
y = pers1(x)
```

```python
import matplotlib.pyplot as plt

plt.plot(x,y)
```

```python
# penentuan parameter

# fitness function

def calculate_fitness(ga_instance, solution, solution_idx):
    output = pers1(solution)
    fitness = -output

    return fitness
```
